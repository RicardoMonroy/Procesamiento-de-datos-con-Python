## uso de merge()

```python
import mysql.connector
import pandas as pd

# Conexión con la base de datos MySQL
cnx = mysql.connector.connect(
    host="127.0.0.1",
    user="root",
    password="admin",
    database="empleados"
)

# Consulta a la tabla de empleados
query_empleados = "select * from empleados"
df_empleados = pd.read_sql(query_empleados, cnx)

# Consulta a la tabla de puestos
query_puestos = "select * from puestos"
df_puestos = pd.read_sql(query_puestos, cnx)

cnx.close()

df_merged = pd.merge(df_empleados, df_puestos, left_on='puestos_idpuestos', right_on='idpuestos')

df_result = df_merged[['nombre', 'fecha_ingreso', 'puesto']]

display(df_result)
```

```python
import mysql.connector
import pandas as pd

# Conexión con la base de datos MySQL
cnx = mysql.connector.connect(
    host="127.0.0.1",
    user="root",
    password="admin",
    database="empleados"
)

# Consulta a la tabla de empleados
query_empleados = "select * from empleados"
df_empleados = pd.read_sql(query_empleados, cnx)

# Consulta a la tabla de líderes
query_lideres = "select * from lideres"
df_lideres = pd.read_sql(query_lideres, cnx)

cnx.close()

df_merged = pd.merge(df_empleados, df_lideres, left_on='lideres_idlideres', right_on='idlideres')

df_result = df_merged[['nombre_x', 'fecha_ingreso', 'nombre_y']]

df_result.columns = ['Empleado', 'Fecha Ingreso', 'Líder']

display(df_result)
```

```python
import mysql.connector
import pandas as pd

# Conexión con la base de datos MySQL
cnx = mysql.connector.connect(
    host="127.0.0.1",
    user="root",
    password="admin",
    database="empleados"
)

# Consulta a la tabla de empleados
query_empelados = "select * from empleados"
query_puestos = "select * from puestos"
query_departamentos = "select * from departamentos"

df_empleados = pd.read_sql(query_empleados, cnx)
df_puestos = pd.read_sql(query_puestos, cnx)
df_departamentos = pd.read_sql(query_departamentos, cnx)

cnx.close()

df_empleados_puestos = pd.merge(df_empleados, df_puestos, left_on='puestos_idpuestos', right_on='idpuestos')
df_completo = pd.merge(df_empleados_puestos, df_departamentos, left_on='departamentos_iddepartamentos', right_on='iddepartamentos')

df_result = df_completo[['nombre', 'fecha_ingreso', 'puesto', 'departamento']]

display(df_result)
```

```python
import mysql.connector
import pandas as pd

# Conexión con la base de datos MySQL
cnx = mysql.connector.connect(
    host="127.0.0.1",
    user="root",
    password="admin",
    database="empleados"
)

# Consulta a la tabla de empleados
query_empelados = "select * from empleados"
query_puestos = "select * from puestos"
query_departamentos = "select * from departamentos"
query_lideres = "select * from lideres"

df_empleados = pd.read_sql(query_empleados, cnx)
df_puestos = pd.read_sql(query_puestos, cnx)
df_departamentos = pd.read_sql(query_departamentos, cnx)
df_lideres = pd.read_sql(query_lideres, cnx)

cnx.close()

df_empleados_puestos = pd.merge(df_empleados, df_puestos, left_on='puestos_idpuestos', right_on='idpuestos')
df_empleados_puestos_departamentos = pd.merge(df_empleados_puestos, df_departamentos, left_on='departamentos_iddepartamentos', right_on='iddepartamentos')
df_completo = pd.merge(df_empleados_puestos_departamentos, df_lideres, left_on='lideres_idlideres', right_on='idlideres')

df_result = df_completo[['nombre_x', 'fecha_ingreso', 'puesto', 'departamento', 'nombre_y']]

df_result.columns = ['Empleado', 'Fecha de ingreso', 'Puesto', 'Departamento', 'Líder']

display(df_result)
```

```python
import mysql.connector
import pandas as pd

# Conexión con la base de datos MySQL
cnx = mysql.connector.connect(
    host="127.0.0.1",
    user="root",
    password="admin",
    database="empleados"
)

# Consulta a la tabla de empleados
query_empelados = "select * from empleados"
query_departamentos = "select * from departamentos"

df_empleados = pd.read_sql(query_empleados, cnx)
df_departamentos = pd.read_sql(query_departamentos, cnx)

cnx.close()

df_empleados_departamentos = pd.merge(df_empleados, df_departamentos, left_on='departamentos_iddepartamentos', right_on='iddepartamentos')

df_salario_promedio = df_empleados_departamentos.groupby('departamento')['sueldo'].mean().reset_index()

df_salario_promedio.columns = ['Departamento', 'Salario Promedioo']

display(df_salario_promedio)
```
